## Part 1: Metropolis-Hastings

**Metropolis-Hastings** targetting **stationary distribution** $p$ based on **proposal distribution** $q$ given below

$$q(\tilde x^{(t)} | x^{(t-1)}) = \left(\frac{2}{3}\right)^{1-{\tilde x^{(t)}}}\left(\frac{1}{3}\right)^{\tilde x^{(t)}} \quad
\textrm{ and } \quad p(x^{(t)}) = \left(\frac{1}{3}\right)^{1-{x^{(t)}}}\left(\frac{2}{3}\right)^{x^{(t)}}$$

has **transition kernel** $K$ of transition probabilities

$K = \left[\begin{array}{cc}\Pr(x^{(t)}=0 \rightarrow x^{(t+1)}=0) & \Pr(x^{(t)}=1 \rightarrow x^{(t+1)}=0)\\\Pr(x^{(t)}=0 \rightarrow x^{(t+1)}=1)& \Pr(x^{(t)}=1 \rightarrow x^{(t+1)}=1)\end{array}\right] = \left[\begin{array}{cl} 2/3 & 1/6 \\ 1/3 & 5/6 \end{array}\right]$

## Part 1B (5%)

### What is the Metropolis-Hastings acceptance probability $\alpha_{mh}$ here?


<!-- You don't have to write this in LaTeX but it's probably easy to do so -->
**$$\alpha_{mh} = 1$$**

$$\alpha_{mh}(\tilde{x}^{(t)} |x^{(t-1)}) = min\left( 1, \frac{p(\tilde{x}^{(t)})q(x^{(t-1)}|\tilde{x}^{(t)})}{p(x^{(t-1)})q(\tilde{x}^{(t)}|x^{(t-1)})}\right) $$
- proposal distribution
$$q(\tilde{x}^{(t)}|x^{(t-1)})=\left(\frac{2}{3}\right)^{1-\tilde{x}^{(t)}}\left(\frac{1}{3}\right)^{\tilde{x}^{(t)}}$$
- target stationary distribution
$$p(x^{(t)})=\left(\frac{1}{3}\right)^{1-x^{(t)}}\left(\frac{1}{3}\right)^{x^{(t)}}$$

$$\alpha_{mh}= min\left( 1, \frac{p(\tilde{x}^{(t)}}{p(x^{(t-1)}}\frac{q(x^{(t-1)}|\tilde{x}^{(t)})}{q(\tilde{x}^{(t)}|x^{(t-1)})}\right) $$


For $x^{(t-1)} = 0$ and $\tilde{x}^{(t)}=1$:

$$\alpha_mh(1|0) = min \left( 1, \frac{(2/3)}{1/3} \cdot \frac{1/3}{2/3}\right) = min (1,1) = 1$$

For $x^{(t-1)} = 1$ and $\tilde{x}^{(t)}=0$:

$$\alpha_mh(0|1) = min \left( 1, \frac{(1/3)}{2/3} \cdot \frac{2/3}{1/3}\right) = min (1,1) = 1$$

### What is the difference between the *acceptance probability,  proposal distribution, transition kernel, stationary distribution and target distribution*?

#### Answer:

<!-- You don't have to use this answer structure template but it might help -->
Proposal Distribution determines how new candidate states are proposed, representing the probability of proposing a new state given the current state, independent of the target distribution. Here, it favors staying in the same state more often.

Acceptance Probability ensures that the Markov chain correctly samples from the target distribution by deciding whether to accept the proposed state or remain in the current state. It adjusts for any asymmetry in the proposal distribution.

The Transition Kernel represents the probability of moving from one state to another, combining both the proposal and acceptance probabilities. It determines how the Markov chain evolves over iterations.

Target Distribution is the distribution we ultimately want to sample from. The Metropolis-Hastings algorithm ensures that this is the stationary distribution.

Stationary Distribution
 If Metropolis-Hastings is implemented correctly, the stationary distribution will be the target distribution.

**To conclude, while the proposal distribution and acceptance probability determine how proposals are made and accepted, the transition kernel governs movement between states. Over time, the chain converges to the stationary distribution, which, in a correctly implemented Metropolis-Hastings algorithm, matches the target distribution.**

## Part 2 (20%): Complete the specification above using PyMC
- Use one beta distribution
- Use one Bernoulli distribution with `shape=272`
- Use four normal distributions for the location $\theta$ parameters, each with `shape=1`
- Use two gamma distributions for the precision $\tau$ parameters, each with `shape=1`
- Use two normal distributions for the two observation columns (as given below)

```python
# If u is a vector (with `shape=272` in this case)
# then `(1-u)*theta_w0 + u*theta_w1` creates a vector of length 272
# which can be used for the mean of the observed `w` data along with `tau=tau_w`
# so long as `shape=1` is used in the specification of `theta_w0` and `theta_w1`
# which makes these random variables a "vector" as opposed to a "scalar"

with pymc.Model() as OldFaithful:
    n = old_faithful.shape[0]

    # One beta distribution
    p = pymc.Beta("p", alpha=1, beta=1)

    # One Bernoulli distribution with `shape=272`
    u = pymc.Bernoulli("u", p=p, shape=n)

    # Four normal distributions for the location $\theta$ parameters, each with `shape=1`
    theta_w0 = pymc.Normal("theta_w0", mu=55, tau=1/5**2)
    theta_w1 = pymc.Normal("theta_w1", mu=85, tau=1/5**2)
    theta_y0 = pymc.Normal("theta_y0", mu=2.0, tau=1/0.5**2)
    theta_y1 = pymc.Normal("theta_y1", mu=4.5, tau=1/0.5**2)

    # Two gamma distributions for the precision $\tau$ parameters, each with `shape=1`
    tau_w = pymc.Gamma("tau_w", alpha=0.5, beta=50)
    tau_y = pymc.Gamma("tau_y", alpha=0.5, beta=0.5)

    # The observed data flows into the model through two entery points
    obs_w = pymc.Normal("obs_w", mu=(1 - u) * theta_w0 + u * theta_w1, tau=tau_w, observed=old_faithful["waiting"])
    obs_y = pymc.Normal("obs_y", mu=(1 - u) * theta_y0 + u * theta_y1, tau=tau_y, observed=old_faithful["eruptions"])

```

## Part 4B (10%): Getting $\theta_w$ and $\theta_y$ from the Joint

$
\begin{align}
\prod_{i=1}^n N(w_i|\theta_{w1},\tau_{w})^{u_i}N(w_i|\theta_{w0},\tau_{w})^{1-u_i} N(y_i|\theta_{y1},\tau_{y})^{u_i}N(w_i|\theta_{y0},\tau_{y})^{1-u_i} &\longleftarrow{} \textrm{likelihood}\\
\times  \prod_{i=1}^n p^{u_i} (1-p)^{1-u_i} \times p(p)p(\theta_{w0})p(\theta_{w1})p(\tau_{w})p(\theta_{y0})p(\theta_{y1})p(\tau_{y}) &\longleftarrow{} \textrm{priors}
\end{align}$


$$\scriptsize
\begin{align*}
p(\theta,\tau|x) &\propto{} p(\theta,\tau,x) = p(x|\theta)p(\theta)p(\tau) \quad (\theta \perp\!\!\perp \tau) \leftarrow \text{independent priors} & p(\theta|x,\theta_0,\tau_0, \tau) &={} \text{N}\left(\frac{\left(\tau_0 \theta_0+\tau\sum_{i=1}^{n}x_{i}\right)}{(\tau_0+n\tau)}, \sigma^{-2}=\tau_0+n\tau \right)\\
&={}  \left[\prod_{i=1}^n\sqrt{\frac{\tau}{2\pi}} e^{-\frac{\tau\left(x_i-\theta\right)^2}{2}}\right] \sqrt{\frac{\tau_0}{2\pi}} e^{-\frac{\tau_0\left(\theta-\theta_0\right)^2}{2}} \frac{\beta ^{\alpha}}{\Gamma(\alpha)} \tau^{\frac{\alpha}{2} -1}e^{-\frac{\beta}{2} \tau} & p(\tau|x, \alpha, \beta, \theta) &={} \text{Gamma}\left(\frac{\alpha}{2}+\frac{n}{2}, \frac{1}{2}\beta+\frac{1}{2}\sum_{i=1}^n\left(x_i-\theta\right)^2 \right)\\{}\\
\end{align*}$$

## Part 4D (10%): Complete Gibbs Sampler


```python
n = 272
G = 100

thetas_w_mu0 = np.array([55,85])
thetas_w = np.zeros((2,G)) + thetas_w_mu0.reshape(2,1)
thetas_w_tau0 = 1/5**2

thetas_y_mu0 = np.array([2.0,4.5])
thetas_y = np.zeros((2,G)) + thetas_y_mu0.reshape(2,1)
thetas_y_tau0 = 1/0.5**2

taus_alpha = 1
taus_beta = np.array([10**2, 1**2])
taus = np.zeros((2,G)) + taus_alpha/taus_beta.reshape(2,1)

us = np.zeros((n,G))
ps = np.zeros(G)+0.5

np.random.seed(42)
waiting_times = np.random.normal(65, 10, n)  # Simulated waiting durations
eruption_times = np.random.normal(3.5, 1, n)  # Simulated eruption durations

for g in range(1,G):

    prob_u1 = (
        stats.norm.pdf(waiting_times, loc=thetas_w[1, g-1], scale=np.sqrt(1 / taus[0, g-1])) *
        stats.norm.pdf(eruption_times, loc=thetas_y[1, g-1], scale=np.sqrt(1 / taus[1, g-1])) * ps[g-1]
    )
    prob_u0 = (
        stats.norm.pdf(waiting_times, loc=thetas_w[0, g-1], scale=np.sqrt(1 / taus[0, g-1])) *
        stats.norm.pdf(eruption_times, loc=thetas_y[0, g-1], scale=np.sqrt(1 / taus[1, g-1])) * (1 - ps[g-1])
    )
    u_probs = prob_u1 / (prob_u1 + prob_u0)
    us[:, g] = stats.bernoulli.rvs(u_probs)

    # Step 2B: Sample p
    alpha_posterior = 1 + np.sum(us[:, g] == 1)
    beta_posterior = 1 + np.sum(us[:, g] == 0)
    ps[g] = stats.beta(alpha_posterior, beta_posterior).rvs()

    # Step 2C: Sample theta_w
    mask_u0 = us[:, g] == 0
    mask_u1 = us[:, g] == 1

    if np.sum(mask_u0) > 0:
        prec_w0 = thetas_w_tau0 + np.sum(mask_u0) * taus[0, g-1]
        tmp_w0 = thetas_w_tau0 * thetas_w_mu0[0] + taus[0, g-1] * np.sum(waiting_times[mask_u0])
        thetas_w[0, g] = stats.norm(loc=tmp_w0 / prec_w0, scale=1 / np.sqrt(prec_w0)).rvs()

    if np.sum(mask_u1) > 0:
        prec_w1 = thetas_w_tau0 + np.sum(mask_u1) * taus[0, g-1]
        tmp_w1 = thetas_w_tau0 * thetas_w_mu0[1] + taus[0, g-1] * np.sum(waiting_times[mask_u1])
        thetas_w[1, g] = stats.norm(loc=tmp_w1 / prec_w1, scale=1 / np.sqrt(prec_w1)).rvs()

    # Step 2D: Sample theta_y
    if np.sum(mask_u0) > 0:
        prec_y0 = thetas_y_tau0 + np.sum(mask_u0) * taus[1, g-1]
        tmp_y0 = thetas_y_tau0 * thetas_y_mu0[0] + taus[1, g-1] * np.sum(eruption_times[mask_u0])
        thetas_y[0, g] = stats.norm(loc=tmp_y0 / prec_y0, scale=1 / np.sqrt(prec_y0)).rvs()

    if np.sum(mask_u1) > 0:
        prec_y1 = thetas_y_tau0 + np.sum(mask_u1) * taus[1, g-1]
        tmp_y1 = thetas_y_tau0 * thetas_y_mu0[1] + taus[1, g-1] * np.sum(eruption_times[mask_u1])
        thetas_y[1, g] = stats.norm(loc=tmp_y1 / prec_y1, scale=1 / np.sqrt(prec_y1)).rvs()

    # Step 2E: Sample tau
    if np.sum(mask_u0) > 0:
        shape_w = taus_alpha / 2 + np.sum(mask_u0) / 2
        rate_w = taus_beta[0] / 2 + np.sum((waiting_times[mask_u0] - thetas_w[0, g])**2) / 2
        taus[0, g] = stats.gamma(shape_w, scale=1/rate_w).rvs()

    if np.sum(mask_u1) > 0:
        shape_y = taus_alpha / 2 + np.sum(mask_u1) / 2
        rate_y = taus_beta[1] / 2 + np.sum((eruption_times[mask_u1] - thetas_y[1, g])**2) / 2
        taus[1, g] = stats.gamma(shape_y, scale=1/rate_y).rvs()

# Step 3: Store results for ArviZ analysis
v1_fit = {
    "p": ps,
    "theta_w0": thetas_w[0],
    "theta_w1": thetas_w[1],
    "theta_y0": thetas_y[0],
    "theta_y1": thetas_y[1],
    "tau_w": taus[0],
    "tau_y": taus[1]
}
```
