# Most Basic Bayesian Interview Question

---

Are you familiar with Bayesian inference?

> Yes!

Well then you must know **the law of total probability**! What's that?

> What proportion of students will pass this class $\Pr(pass)$ if you know $\Pr(pass|quick)$ as well as the distribution of $quick$?

Now what proportion of students who will pass the class answer this question quickly?

> The structure of this question is **_the most basic Bayesian interview question_**, but it requires some assumptions...

# Bayesian Terminology
---

## Bayes Theorem

\begin{align*}
\overset{posterior}{p(\theta|x)} &={} \frac{\overset{likelihood}{p(x|\theta)}\times \overset{prior}{p(\theta)}}{\overset{\text{evidence}}{p(x)}}\\
&={} \frac{{p(x, \theta)}}{\underset{\text{marginal likelihood}}{\displaystyle \int p(x|\theta)p(\theta) d\theta}} \textrm{ or }
\frac{{p(x, \theta)}}{\underset{\text{law of total probability}}{\displaystyle \sum_{k} p(x|\theta_k)p(\theta_k)}} =
\frac{\overset{joint}{p(x, \theta)}}{\underset{\text{normalizing constant}}{c}}
\end{align*}

### _The posterior is always proportional to the joint_

$$\large p(\theta|x) = (1/c)\times p(\theta,x) \propto{} p(\theta,x)$$

**_and the posterior is always a proper distribution if the prior is_**

# Example
---
\begin{align*}
\text{Likelihood} && p(x|\theta,n) &={} C^n_k\theta^{x}(1-\theta)^{n-x}\\
&&& \quad\; [\text{exponential family distribution if $n$ is known}]\\
&&& \quad\; [\text{What is this distribution? What is $C^n_k$?}]\\\\
\text{Prior} && p(\theta|\alpha,\beta) &={} \frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\Gamma(\beta)}\theta^{\alpha-1}(1-\theta)^{\beta-1} \\
&&& \quad\; [\text{$\alpha$ and $\beta$ are called }\textbf{hyperparameters}]\\
&&& \quad\; [\text{What is this distribution? What is $\Gamma(\cdot)$?}]\\\\
\text{Posterior} && p(\theta|x,n,\alpha,\beta) &={} \overset{\text{Bayes Theorem}}{\frac{p(x|\theta)p(\theta)}{p(x)}} = \frac{C^n_k\theta^{x}(1-\theta)^{n-x}  \frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\Gamma(\beta)}\theta^{\alpha-1}(1-\theta)^{\beta-1} }{p(x)} \\
&&& = \;\texttt{What distribution parameterized how?}
\end{align*}

# **Homework 1 Solution**

## **Q1**

Given:

*   $P(\text{Pass}) = 0.9$
*   $P(\text{Not Pass}) = 0.1$
*   $P(\text{Quick|Pass}) = 0.6$
*   $P(\text{Quick|Not Pass}) = 0.3$

Want to find: $P(\text{Pass}|\text{Quick})$

By Bayes' Theorem:

$$P(\text{Pass}|\text{Quick}) = \frac{P(\text{Quick}|\text{Pass)}P\text{(Pass)}}{P( \text{Quick)}} $$

Using the law of total probability

$$P(\text{Quick}) = P\text{(Quick|Pass)}P\text{(Pass)}+ P\text{(Quick|Not Pass)}P\text{(Not Pass)}$$

$$P\text{(Quick)} = (0.6)(0.9) + (0.3)(0.1) = 0.54 + 0.03 = 0.57$$


$$P \text{(Pass|Quick)} = \frac{(0.6)(0.9)}{0.57} = \frac{0.54}{0.57} \approx 0.947$$

So about 94.7% of students who answer quickly will pass the class


## **Q2**

Given:

\begin{align*}
\text{Likelihood} && p(x|\theta,n) &={} \texttt{Multinomial}\\\\
\text{Prior} && p(\theta|\boldsymbol{\alpha}) &={} \texttt{Dirichlet} \\
\text{Posterior} && p(\theta|x,n,\boldsymbol{\alpha}) &={} \overset{\text{Bayes Theorem}}{\frac{p(x|\theta)p(\theta)}{p(x)}} = \; \texttt{What?}\\
&&& \quad\quad\quad\quad\quad\; = \; \texttt{What distribution?}\\
&&& \quad\quad\quad\quad\quad\quad\;\;\, \texttt{With what parameterization?}\\
\end{align*}

Likelihood is multinomial

$$p(x|\theta, n) = \frac{n!}{x_1!x_2!\cdots x_k!} \prod_{i=1}^{k} \theta_i^{x_1}$$

Prior is Dirichlet

$$p(\theta|a) \propto \prod_{i=1}^{k} \theta_i^{a_1-1}$$

Posterior

$$p(\theta|x, n, a) \propto p(x|\theta, n) p(\theta|a)$$

$$p(\theta|x, n, a) \propto \prod_{i=1}^{k} \theta_i^{x_1} \prod_{i=1}^{k} \theta_i^{a_1-1}$$

$$p(\theta|x, n, a) \propto \prod_{i=1}^{k} \theta_i^{x_1+a_1-1}$$

Posterioir is a Dirichlet distribution

$$p(\theta|x, n, a) = \text{Dirichlet } (a+x)$$

where the parameters are:
$$a' = a+x = (a_1 + x_1, a_2+x_2 \dots a_k+x_k)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multinomial, dirichlet
import seaborn as sns
!pip install python-ternary==1.0.6
import ternary
```

```python
### Q3: Visualizing Multinomial and Dirichlet Distributions


# Set random seed for reproducibility
np.random.seed(42)

# 1. Create a multinomial distribution and sample
n_trials = 100
probabilities = [0.4, 0.35, 0.25]  # Probabilities for the 3 categories
sample = multinomial.rvs(n_trials, probabilities)

# 2. Visualize the multinomial distribution and sample
plt.figure(figsize=(8, 4))
sns.barplot(x=["Category 1", "Category 2", "Category 3"], y=probabilities, color="skyblue")
plt.title("Multinomial Distribution")
plt.ylabel("Probability")
plt.show()

plt.figure(figsize=(8, 4))
sns.barplot(x=["Category 1", "Category 2", "Category 3"], y=sample / n_trials, color="orange")
plt.title("Multinomial Sample")
plt.ylabel("Proportion")
plt.show()

# 3. Define Dirichlet prior
alpha_prior = np.array([2, 2, 2])  # Symmetric prior for simplicity

# 4. Update the prior to compute the posterior
alpha_posterior = alpha_prior + sample

# 5. Visualize the prior and posterior Dirichlet distributions using ternary plots
# Prepare the data for plotting
scale = 100
prior_points = {}
posterior_points = {}

for i in range(scale + 1):
    for j in range(scale - i + 1):
        k = scale - i - j
        point = (i / scale, j / scale, k / scale)
        prior_points[point] = dirichlet.pdf(point, alpha_prior)
        posterior_points[point] = dirichlet.pdf(point, alpha_posterior)

# Normalize for better visualization
max_prior = max(prior_points.values())
max_posterior = max(posterior_points.values())
prior_points = {key: value / max_prior for key, value in prior_points.items()}
posterior_points = {key: value / max_posterior for key, value in posterior_points.items()}

# Plot Dirichlet prior
fig, ax = plt.subplots(figsize=(8, 8))
figure, tax = ternary.figure(ax=ax, scale=1.0)
tax.heatmap(prior_points, cmap="Blues", style="hexagonal", colorbar=True)
tax.boundary()
tax.set_title("Dirichlet Prior", fontsize=16)
tax.left_axis_label("Component 1", fontsize=12)
tax.right_axis_label("Component 2", fontsize=12)
tax.bottom_axis_label("Component 3", fontsize=12)
plt.show()

# Plot Dirichlet posterior
fig, ax = plt.subplots(figsize=(8, 8))
figure, tax = ternary.figure(ax=ax, scale=1.0)
tax.heatmap(posterior_points, cmap="Reds", style="hexagonal", colorbar=True)
tax.boundary()
tax.set_title("Dirichlet Posterior", fontsize=16)
tax.left_axis_label("Component 1", fontsize=12)
tax.right_axis_label("Component 2", fontsize=12)
tax.bottom_axis_label("Component 3", fontsize=12)
plt.show()
```
